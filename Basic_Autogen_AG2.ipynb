This Python code uses the **AutoGen (ag2)** and **Groq (groq)** packages to create a **conversational AI system** that summarizes team updates and provides project timeline recommendations.

It uses Large Language Models (LLMs) to achieve this by setting up two agents:

1.   a **Summarizer** Agent
2.   a **Recommender** Agent

The agents collaborate in a simulated **group chat** to process input data (team updates) and generate meaningful outputs (summaries and recommendations).

The setup uses asynchronous loops suitable for Jupyter notebooks with the nest_asyncio package.

AutoGen is most effective for structured, collaborative problem-solving where agents play distinct but interdependent roles.

```python
#Imports necessary modules
from google.colab import userdata
from autogen import ConversableAgent
from pydantic import BaseModel
```

```python
# Set your Groq API key
GROQ_API_KEY = userdata.get('GROQ_API_KEY')
```

```python
# Manage workflow state variables to ensure proper data structure and validation.
class WorkflowState(BaseModel):
    team_updates: str = ""
    task_summary: str = ""
    decision_support: str = ""
```

```python
# Define LLM configuration for Groq models
llm_config = {
    "config_list": [
        {
            "model": "llama3-8b-8192",
            "api_key": GROQ_API_KEY,
            "api_type": "groq"
        }
    ]
}
```

```python
# Summarizer Agent
summarizer_agent = ConversableAgent(
    name="summarizer_agent",
    llm_config=llm_config,
    system_message="You are an AI assistant tasked with summarizing team updates."
)
```

```python
# Recommender Agent
recommender_agent = ConversableAgent(
    name="recommender_agent",
    llm_config=llm_config,
    system_message="You provide recommendations for project timeline adjustments based on task summaries."
)
```

```python
# Using Autogen to orchestrate workflow in conversational style
from autogen import GroupChat, GroupChatManager

# Team updates input
team_updates = """
- Alice completed the UI design and is working on feedback integration.
- Bob fixed the critical bug and is now optimizing backend queries.
- Charlie is struggling with API integration and needs assistance.
- Dave is on track with the deployment pipeline improvements.
"""

# Initialize GroupChat with agents
groupchat = GroupChat(
    agents=[summarizer_agent, recommender_agent],
    speaker_selection_method="auto",
    messages=[]
)

# Define termination condition
def is_termination_msg(message):
    return "DONE!" in message.get("content", "").upper()

# Create GroupChatManager
manager = GroupChatManager(
    name="workflow_manager",
    groupchat=groupchat,
    llm_config=llm_config,
    is_termination_msg=is_termination_msg
)

# Start the conversation
chat_result = summarizer_agent.initiate_chat(
    recipient=manager,
    message=f"Summarize the following team updates:\n{team_updates}"
)

# Print the chat history
print(chat_result.chat_history)
```
